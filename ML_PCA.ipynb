#### Коефіцієнт детермінаії:

```python
y_mean=Y_train.mean()
```

```python
n = 200
Qr = float(np.dot(np.dot(np.transpose(b), Xt), Y_train)) - n*(y_mean**2)
Q = float(np.dot(np.transpose(Y_train), Y_train)) - n*(y_mean**2)
```

```python
#Коефіцієнт кореляції
R = sqrt(Qr/Q)
R
```

```python
#Коефіцієнт детермінації
R2 = R**2
R2
```
