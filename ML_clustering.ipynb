```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option('display.max_rows', None)
df=pd.read_csv("MN-03-Hryhorenko.csv", sep = ';', header = None)
df.rename(columns = {0:'x1',1:'x2',2:'x3',3:'x4',4:'x5',5:'x6',6:'x7',7:'y'}, inplace = True)
```

```python
X = df[['x1','x2','x3','x4','x5','x6','x7']]
```

### Розбиття методом  K-means на два кластери:

```python
ModelKMeans2 = KMeans(n_clusters=2).fit(X)
```

```python
X['cluster_kmeans2'] = ModelKMeans2.labels_
X
```

```python
fig, ax = plt.subplots(2,2)
fig.set_size_inches(10, 7)
ax[0][0].set_xlabel('x1')
ax[0][0].set_ylabel('x2')
ax[0][0].scatter(X['x1'], X['x2'], c= X['cluster_kmeans2'], 
           alpha=0.5, edgecolors='none')
ax[0][1].set_xlabel('x3')
ax[0][1].set_ylabel('x4')
ax[0][1].scatter(X['x3'], X['x4'], c= X['cluster_kmeans2'], 
           alpha=0.5, edgecolors='none')
ax[1][0].set_xlabel('x5')
ax[1][0].set_ylabel('x6')
ax[1][0].scatter(X['x5'], X['x6'], c= X['cluster_kmeans2'], 
           alpha=0.5, edgecolors='none')
ax[1][1].set_xlabel('x6')
ax[1][1].set_ylabel('x7')
ax[1][1].scatter(X['x6'], X['x7'], c= X['cluster_kmeans2'], 
           alpha=0.5, edgecolors='none')
fig.show()
```

```python
X.drop('cluster_kmeans2', axis = 1, inplace =True)
```

### Розбиття методом K-means на три кластери:
